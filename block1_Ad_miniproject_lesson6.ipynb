```python
# Читаем через csv

path             = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-o-bogdanova/shared/homeworks/python_ds_miniprojects/6_gff/gff.tsv'
fly_genome = pd.read_csv(path, comment='"', sep='\s+', header=1, names=[ "seq_id", "source", "type", "start", "end", "score", "strand", "phase","attributes"])
fly_genome
```

```python
# Какое значение атрибута Parent самое частое?
pattern = re.compile('Parent=') 
fly_genome['Parent'] = [re.split(pattern, i)[1] for i in fly_genome.attributes] #Разделяем строку по паттерну
fly_genome \
    .groupby('Parent', as_index = False) \
    .agg({'seq_id' : 'count'}) \
    .sort_values('seq_id', ascending = False) \
    .head(1)
```

```python
# Какое самое частое значение в колонке type?
fly_genome \
    .groupby('type', as_index = False) \
    .agg({'seq_id' : 'count'}) \
    .sort_values('seq_id', ascending = False) \
    .head(1)
```
