```python
# Импорт библиотеки
import pandas as pd
```

```python
# Читаем датасет
bookings = pd.read_csv('2_bookings.csv', sep=';')
```

```python
# Выводим 7 строк таблицы
bookings_head = bookings.head(n=7)
bookings_head
```

```python
# Размер таблицы
bookings.shape
```

```python
# Типы данных в таблице
bookings.dtypes
```

```python
# Переименновывание столбцов
def replace_space_and_capital_letters(name):
    new_name = name.lower().replace(' ', '_', )
    return new_name
bookings = bookings.rename(columns=replace_space_and_capital_letters)
bookings.head()
```

```python
# Пользователи из каких стран совершили наибольшее число успешных бронирований?
country_success_client = (bookings
                        .query("is_canceled == 0")
                        .groupby('country', as_index = False)
                        .agg({'is_canceled':'count'})
                        .sort_values('is_canceled', ascending = False)
                     )
country_success_client
```

```python
# Расчет среднего времени бронирования отелей типа City Hotel и Resort Hotel с точностью до 2 знаков после точки.
stays_total_nights = (bookings.groupby('hotel', as_index = False)
                   .agg({'stays_total_nights':'mean'})
                   .round(decimals = 2))
stays_total_nights
```

```python
# Подсчитать количество случаев, когда тип номера, присвоенного клиенту (assigned_room_type) отличается от изначально забронированного (reserved_room_type)
different_room_type = (bookings
                        .query("assigned_room_type!=reserved_room_type")
                     .shape[0])
different_room_type
```

```python
# Какой самый популярный месяц в 2016 и в 2017 году?
count_arrivals_by_month = (bookings
                         .groupby(['arrival_date_year','arrival_date_month'], as_index=False)
                         .agg({'is_canceled':'count'})
                         .sort_values(['arrival_date_year', 'is_canceled'], ascending=False))
count_arrivals_by_month
```

```python
#В какой месяц бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017? 
count_canc_arriv_month_1 = (bookings
                        .query("is_canceled==1 and hotel=='City Hotel'")
                         .groupby(['arrival_date_year','arrival_date_month'], as_index=False)
                         .agg({'is_canceled':'count'})
                            .sort_values(['arrival_date_year', 'is_canceled'], ascending=False)
                            #.groupby('arrival_date_year', as_index=False)
                             #.nlargest(n = 5, columns = 'is_canceled')
                            )
count_canc_arriv_month_1
```

```python
# Другой способ определить в какой месяц бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017?
count_canc_arriv_month_2 = (bookings
                        .query("is_canceled == 1 and hote l== 'City Hotel'")
                         .groupby('arrival_date_year', as_index=False)['arrival_date_month']
                          .sum())
count_canc_arriv_month_2
```

```python
#  . Какая из трёх колонок adults, children и babies имеет наибольшее среднее значение?
mean_adults = bookings['adults'].mean()
mean_children = bookings['children'].mean()
mean_babies = bookings['babies'].mean()
mean_adults, mean_children, mean_babies
```

```python
# Создайте колонку total_kids, объединив столбцы children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?
bookings = bookings.assign(total_kids = bookings['children']+bookings['babies'])
mean_total_kids_by_hotel =  (bookings
                             .groupby('hotel', as_index = False)
                             .agg({'total_kids':'mean'})
                             .round(decimals=2)
                            .nlargest(n = 1, columns = 'total_kids'))
mean_total_kids_by_hotel 
```

```python
# Посмотрим, как Churn rate (отток, коэффициент оттока) связан с наличием детей у клиентов!
#Создайте колонку has_kids: 
# - True, если клиент указал хотя бы одного ребенка (total_kids);
# - False, если детей 0.

bookings['has_kids'] = bookings['total_kids']>0

# Проверяем, среди какой группы пользователей показатель оттока выше
       #Находим общее количество клиентов, клиентов с детьми и без детей
total_clients = bookings.shape[0]
clients_with_kid = (bookings
                         .query("has_kids == True")
                         .shape[0])
clients_without_kid = (bookings
                         .query("has_kids == False")
                         .shape[0]) 

        #Находим общее количество клиентов, клиентов с детьми и без детей, которые ОТМЕНИЛИ бронь
canc_total_cli =  (bookings
                         .query("is_canceled==1")
                         .shape[0])    
canc_cli_with_kid = (bookings
                         .query("has_kids == True and is_canceled==1")
                         .shape[0])
canc_cli_without_kid = (bookings
                         .query("has_kids == False and is_canceled==1")
                         .shape[0])
        #Вычисляем Churn rate
churn_rate_total = canc_total_cli / total_clients * 100
churn_rate_with_kid = canc_cli_with_kid / clients_with_kid * 100
churn_rate_without_kid = canc_cli_without_kid / clients_without_kid * 100

#  Проводим проверку вычислений
        #Должно сойтись кол-во клиентов с/без детей с общим кол-вом клиентов, аналогично с отказами от брони.
if total_clients == clients_with_kid + clients_without_kid and canc_total_cli == canc_cli_with_kid + canc_cli_without_kid:
            # Коэффициент оттока по всем клиентам должен находится в промежутке между рассчитываемыми частностями: с/без детей.
    if churn_rate_with_kid >= churn_rate_total >= churn_rate_without_kid or churn_rate_with_kid <= churn_rate_total<= churn_rate_without_kid:
        print('Проверка пройдена!')
        print( 'churn_rate_with_kid:', round(churn_rate_with_kid, 2), 'churn_rate_without_kid:', round(churn_rate_without_kid, 2))
    else:     print("Error")
else:     print("Error")
    
```
